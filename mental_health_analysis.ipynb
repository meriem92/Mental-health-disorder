```python
import pandas as pd
import plotly.express as px
import pycountry_convert as pc
import matplotlib.pyplot as plt

# Load Data
df=pd.read_csv('Mental health Depression disorder Data.csv')
```

DATA CLEANING

```python
df=df.drop('index',axis=1)
df=df.dropna()
df=df[df['Entity']!='World']
```

```python
#Fix all column types
df['Entity']=df['Entity'].astype('string')
df['Code']=df['Code'].astype('string')
df['Year']=pd.to_numeric((df['Year']),downcast='integer',errors='coerce').astype('Int64')
df['Schizophrenia (%)']=pd.to_numeric(df['Schizophrenia (%)'],errors='coerce').round(3)
df['Schizophrenia (%)'].apply(lambda col: format(col,'3f'))

df['Bipolar disorder (%)']=pd.to_numeric(df['Bipolar disorder (%)'],errors='coerce').round(3)
df['Bipolar disorder (%)'].apply(lambda col: format(col,'3f'))

df['Eating disorders (%)']=pd.to_numeric(df['Eating disorders (%)'],errors='coerce').round(3)
df['Eating disorders (%)'].apply(lambda col: format(col,'3f'))

df['Anxiety disorders (%)']=pd.to_numeric(df['Anxiety disorders (%)'],errors='coerce').round(3)
df['Anxiety disorders (%)'].apply(lambda col: format(col,'3f'))

df['Drug use disorders (%)']=pd.to_numeric(df['Drug use disorders (%)'],errors='coerce').round(3)
df['Drug use disorders (%)'].apply(lambda col: format(col,'3f'))

df['Depression (%)']=pd.to_numeric(df['Depression (%)'],errors='coerce').round(3)
df['Depression (%)'].apply(lambda col: format(col,'3f'))

df['Alcohol use disorders (%)']=pd.to_numeric(df['Alcohol use disorders (%)'],errors='coerce').round(3)
df['Alcohol use disorders (%)'].apply(lambda col: format(col,'3f'))
```

```python
df.groupby(['Entity','Code','Year']).sum()
```

```python
df['Entity'].unique()
```

```python
df['Entity']=df['Entity'].apply(lambda x:'South America' if x=='Andean Latin America' else x)
df['Entity']=df['Entity'].apply(lambda x:'Ivory Coast' if x=="Cote d'Ivoire" else x)
df['Entity']=df['Entity'].apply(lambda x:'Congo' if x=="Democratic Republic of Congo" else x)
df['Entity']=df['Entity'].apply(lambda x:'Micronesia' if x=="Micronesia (country)" else x)
df['Entity']=df['Entity'].apply(lambda x:'East Timor' if x=="Timor" else x)
```

```python
def country_to_continent(country_name):
    country_alpha2 = pc.country_name_to_country_alpha2(country_name)
    country_continent_code = pc.country_alpha2_to_continent_code(country_alpha2)
    country_continent_name = pc.convert_continent_code_to_continent_name(country_continent_code)
    return country_continent_name
```

```python
df['Continent']=df['Entity'].apply(lambda x: country_to_continent(x) if x!='East Timor'else 'Asia')
```

```python
df['Continent'].unique()
```

```python
df. groupby(['Entity','Code','Year','Continent']).sum()
```

```python
df_Continent=df[['Continent','Schizophrenia (%)','Bipolar disorder (%)','Eating disorders (%)','Anxiety disorders (%)','Drug use disorders (%)','Depression (%)','Alcohol use disorders (%)']].copy()
```

```python
df_Continent=df_Continent.groupby('Continent').sum().reset_index()
```

```python
# Mental Disorders by Continent in 2017
df_Continent_2017=df.copy()
df_Continent_2017=df_Continent_2017[df_Continent_2017['Year']==2017]
df_Continent_2017=df_Continent_2017[['Continent','Schizophrenia (%)','Bipolar disorder (%)','Eating disorders (%)','Anxiety disorders (%)','Drug use disorders (%)','Depression (%)','Alcohol use disorders (%)']].copy()
```

```python
df_Continent_2017=df_Continent_2017.groupby('Continent').sum().reset_index()
```
